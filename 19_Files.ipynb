```python
#텍스트 파일이 용량이 클 경우 파일을 한줄씩 읽고 출력 (variable.readline()module)
f=open('/content/sample_data/test/crime_punishment_chapter_one_sample.txt','rt')
line_num = 1
line = f.readline()

while line:
  print('%d %s' %(line_num, line), end='') #end='' 코드라인은 한 줄 삽입에 대한 명령어
  line = f.readline()
  line_num += 1  #사칙연산을 사용하여 좌측 변수와 우측정수를 함께 계산하여 동일한 변수에 할당할 때 사용하는 코드라인 variable += number
f.close()
```

```python
#텍스트 파일의 끝까지 한줄씩 읽어 각 줄을 요소로 하는 리스트 리턴 (variable.readlines()module) 

f=open('/content/sample_data/test/crime_punishment_chapter_one.txt', 'rt')
lines = f.readlines() #텍스트의 모든 내용을 한줄씩 '한꺼번'에 꺼내 읽기
for line_number, line in enumerate(lines): #enumerate -을 하나씩 열거하다
  print('%d %s' %(line_number+1, line), end='')
f.close()
```

```python
#텍스트 파일에 한줄씩 쓰기 (variable.writelines()module)

count = 1
data = []
print('파일에 내용을 저장하려면 내용을 입력하지 말고 ENTER를 누르세요')

while True:
  text = input('[%d] 파일에 저장할 내용을 입력하세요:'%count) #
  if text =='': #만일 입력한 내용이 공란이면
    break #중단하라
  data.append(text+'\n') #줄바꿈 기호 \n을 추가한 문자열을 리스트자료 data에 추가한다. 
  count +=1 #이후 1씩 중가

f=open('lines.txt','wt')
f.writelines(data) #데이터의 모든 요소 (리스트자료로 한줄씩 입력한 것)를 파일에 기록
f.close

#책과 동일하게 했는데, 0byte로 아무런 기록이 파일에 되어 있지 않았음. 수차레. 
```

```python
#파일을 열고 자동으로 닫기 (with ~ as): 파일에 대한 처리가 끝나면 자동으로 파일을 닫아준다.

with open('/content/sample_data/Temp/crime_punish_paragraph.txt', 'rt') as f:
  for line_number, line in enumerate(f.readlines()):
    print('%d %s' %(line_number+1, line), end='')
```

```python
#파일의 특정 부분만 복사하기 (seek, read, write)

loc = 105 #파일을 읽는 위치 지정하는데 (i.e., 파일의 105 byte 위치부터) 
size = 500 #파일을 읽을 크기를 저장하는데 (i.e., 파일을 500 byte 읽어서)

f = open('/content/sample_data/Temp/crime_punish_paragraph.txt', 'rt')
c = open('/content/sample_data/Temp/crime_punish_selected.txt', 'wt')

f.seek(loc)
data = f.read(size)
c.write(data)

f.close()
c.close()
```
